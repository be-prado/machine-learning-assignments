### This code classify star signal data to determine whether the signal came from a pulsar or not. We use an l^2 penalty hinge loss objective function and we implement gradient descent and stochastic gradient descent to find the classification.

Author: Bernardo Bianco Prado

Date: 12/26/2021

More information about the dataset can be found at archive.ics.uci.edu/ml/datasets/HTRU2

```python
# IMPORT LIBRARIES AND PLOT THE DATA

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

x = np.load("pulsar_features.npy")
y = np.load("pulsar_labels.npy")

d,n = np.shape(x)
_lambda = 0.001
#print(y[0,2000])

negInd = y == -1
posInd = y == 1
plt.scatter(x[0, negInd[0, :]], x[1, negInd[0, :]], color='b', s=0.3)
plt.scatter(x[0, posInd[0, :]], x[1, posInd[0, :]], color='r', s=0.3)
plt.figure(1)
#print(np.size(x[:,0])+1)
plt.show()
```

```python
# USEFUL FUNCTIONS

# compute subgradiend u_i
def u(x_i,y_i,_w,_b):
    _subgrad = np.zeros(d+1)
    if 1-y_i*(np.dot(_w,x_i) + _b) > 0:
        _subgrad[0] = -y_i/n
        _subgrad[1:] =(-y_i*x_i +_lambda*_w)/n
        return _subgrad
    else:
        _subgrad[1:] =(_lambda*_w)/n
        return _subgrad

```

```python
# GRADIENT DESCENT

print("_______________BEGINNING GRADIENT DESCENT_________________\n")

iterations = 10
# step size
eta = 100
# weight vector
w = np.zeros(d)
# bias
b = 0
# augmented weight vector
w_bias = np.zeros(d+1)
# vector of objective function J for each iteration
J_vector = np.zeros(iterations)

# gradient descent loop
for j in range(iterations):
    # update eta
    eta = 100/(j+1)
    # compute subgradient
    subgrad = np.zeros(d+1)
    for i in range(n):
        # add the subgradient u_i corresponding to instance x_i
        subgrad += u(x[:,i],y[0,i],w,b)
    # update augmented w
    w_bias = w_bias - eta*subgrad
    # split w and b
    w = w_bias[1:]
    b = w_bias[0]
    #print("subgrad:",subgrad,"\nw:", w_bias, "\neta:",eta,"\niteration:",j+1,"\n\n\n")
    
    # compute objective function at this iteration
    J = 0
    for i in range(n):
        t = np.dot(x[:,i],w_bias[1:]) + w_bias[0]
        _max = np.maximum(0, 1-y[0,i]*t)
        J += (_max+_lambda*np.dot(w_bias[1:],w_bias[1:])/2)/n
    print("J after iteration", j+1, ":", J)
    # store computed objective function
    J_vector[j] = J

print("\nHyperplane parameters: \n", "\tb = ", w_bias[0], "| w_1 = ",w_bias[1], "| w_2 = ",w_bias[2])

# compute margin = 1/||w||
margin = 1/np.linalg.norm(w)

print("\nMargin:", margin)
    
print("\nBest value of J:", np.min(J_vector))
```

```python
# PLOT LEARNED LINE

fig, axs = plt.subplots(1,2,constrained_layout=True)


negInd = y == -1
posInd = y == 1
# plot of learned line with data points
axs[0].plot(x[0,:], (-b - w[0]*x[0,:])/w[1])
axs[0].scatter(x[0, negInd[0, :]], x[1, negInd[0, :]], color='b', label = 'Not Pulsar', s=0.3)
axs[0].scatter(x[0, posInd[0, :]], x[1, posInd[0, :]], color='r',label = 'Pulsar', s=0.3)
#axs[0].figure(1)
axs[0].set_title('Pulsar Data with Learned Line')
axs[0].set_xlabel('Mean of the integrated profile')
axs[0].set_ylabel('Excess kurtosis of the DM-SNR curve')
axs[0].legend()

# plot of J for each iteration
axs[1].scatter(np.arange(1,11), J_vector, color = 'green')
axs[1].set_title('Learned J per Iteration')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('J')

fig.suptitle('Results for Gradient Descent', fontsize=16)

plt.savefig("p3_gradient_descent.png")

plt.show()
```

```python
# STOCHASTIC GRADIENT DESCENT

print("______________BEGINNING STOCHASTIC GRADIENT DESCENT____________\n")

# weight vector
sw = np.zeros(d)
# bias
sb = 0
# augmented weight vector
sw_bias = np.zeros(d+1)
# vector of objective function J for each iteration
sJ_vector = np.zeros(iterations)

# stochastic gradient descent loop
for j in range(iterations):
    # update stepsize
    eta = 100/(j+1)
    # randomly order indices of instances
    order = np.random.permutation(n)
    # loop through randomly ordered instances to compute u_{order[i]} and update weight vector
    for i in range(n):
        # compute u_i
        subgrad = u(x[:,order[i]],y[0,order[i]],sw,sb)
        # update weight vector according to instance x_{order[i]}
        sw_bias = sw_bias - eta*subgrad
        # splite w and b
        sw = sw_bias[1:]
        sb = sw_bias[0]
    
    # compute objective function at this iteration
    sJ = 0
    for i in range(n):
        t = np.dot(x[:,i],sw_bias[1:]) + sw_bias[0]
        _max = np.maximum(0, 1-y[0,i]*t)
        sJ += (_max+_lambda*np.dot(sw_bias[1:],sw_bias[1:])/2)/n
    print("J after iteration", j+1, ":", sJ)
    # store objective function at this iteration
    sJ_vector[j] = sJ
    
        
print("\nHyperplane parameters: \n", "\tb = ", sw_bias[0], "| w_1 = ",sw_bias[1], "| w_2 = ",sw_bias[2])

# compute margin = 1/||w|| (the distance between w^T x + b = 0 and w^T x + b = 1)
smargin = 1/np.linalg.norm(sw)

print("\nMargin:", smargin)
    
print("\nBest value of J:", np.min(sJ_vector))
```

```python
# PLOT LEARNED LINE

fig2, axs2 = plt.subplots(1,2,constrained_layout=True)


negInd = y == -1
posInd = y == 1
# plot of learned line with data points
axs2[0].plot(x[0,:], (-sb - sw[0]*x[0,:])/sw[1])
axs2[0].scatter(x[0, negInd[0, :]], x[1, negInd[0, :]], color='b', label = 'Not Pulsar', s=0.3)
axs2[0].scatter(x[0, posInd[0, :]], x[1, posInd[0, :]], color='r',label = 'Pulsar', s=0.3)
#axs[0].figure(1)
axs2[0].set_title('Pulsar Data with Learned Line')
axs2[0].set_xlabel('Mean of the integrated profile')
axs2[0].set_ylabel('Excess kurtosis of the DM-SNR curve')
axs2[0].legend()

# plot of J for each iteration
axs2[1].scatter(np.arange(1,11), sJ_vector, color = 'green')
axs2[1].set_title('Learned J per Iteration')
axs2[1].set_xlabel('Iteration')
axs2[1].set_ylabel('J')

fig2.suptitle('Results for Stochastic Gradient Descent', fontsize=16)

plt.savefig("p3_stochastic_gradient_descent.png")

plt.show()
```
