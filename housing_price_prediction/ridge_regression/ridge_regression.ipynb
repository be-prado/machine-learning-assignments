```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
np.random.seed(0)
```

```python
Xtrain = np.load("housing_train_features.npy")
Xtest = np.load("housing_test_features.npy")
ytrain = np.load("housing_train_labels.npy")
ytest = np.load("housing_test_labels.npy")

feature_names = np.load("housing_feature_names.npy", allow_pickle=True)
#print("First feature name: ", feature_names[0])
#print("Lot frontage for first train sample:", Xtrain[0,0])
```

```python
# SPHERE THE DATA

d,n =np.shape(Xtrain)
# create sphered feature matrix
spheredX = np.copy(Xtrain)
# create feature empirical mean and variance vectors
mean = np.zeros(d)
sd = np.ones(d)

for i in range(d):
    # empirical mean of feature i
    mean[i] = np.mean(Xtrain[i])
    # empirical variance of feature i
    sd[i] = np.sqrt(np.var(Xtrain[i]))
    # sphered values of i-th feature of the x instances
    spheredX[i] = (spheredX[i] - mean[i]*np.ones(n))/sd[i]

#print(np.var(spheredX[0]), np.mean(spheredX[5]))
#print(mean)

smean = np.zeros(d)
svar = np.zeros(d)
for i in range(d):
    smean[i] = np.mean(spheredX[i])
    svar[i] = np.var(spheredX[i])

# print means and variances after sphering
print("max mean:",np.max(smean),"| min mean:", np.min(smean),"\nmax variance:", \
                  np.max(svar[i]),"| min variance:",np.min(svar[i]))
```

```python
# PLOT SAMPLE MEANS AND VARIANCES OF THE FEATURES

# vector [0,1,...,d-1]
x = np.arange(0,d)

fig, axs = plt.subplots(2, 1,constrained_layout=True)

# plot sample mean of features
#plt.yscale('log')
axs[0].plot(x,mean,'ro')
axs[0].set_title('Sample Mean')
axs[0].set_xlabel('Feature index')
axs[0].set_ylabel('Mean')

# plot sample variance of features
axs[1].plot(x,sd,'ro',color = 'b')
axs[1].set_title('Sample Variance')
axs[1].set_xlabel('Feature index')
axs[1].set_ylabel('Variance')

# save plot
plt.savefig("mean_var.png")

plt.show()
```

```python
# USEFUL FUNCTIONS

# function to compute minimizing vector w
def compute_w(_lambda,sXb,y):
    #print(_lambda)
    # Create identity matrix with 0 on the first entry to not account for w_0 in the
    # penalty term.
    modified_id = np.identity(d+1)
    modified_id[0,0] = 0

    # Compute matrix [Phi^T * Phi - lambda * modified_id]^{-1}.
    # This takes into account the bias term in the solution for the vector w without
    # considering w_0 in the penalty term.
    matrix = np.linalg.inv(np.matmul(sXb, np.transpose(sXb)) + _lambda * modified_id)
    #print("lambda:", _lambda,"norm:",np.linalg.norm(matrix))

    # vector Phi*y
    vector = np.matmul(sXb,y)
    
    # return w = [Phi^T * Phi - lambda * modified_id]^{-1} * Phi * y
    return np.matmul(matrix,vector)


# compute MSE of the found w
def mse(w,x,y):
    # number of instances
    n = np.size(y)
    # initialize MSE
    _MSE = 0
    # add MSE term for each instance
    for i in range(n):
        # (w^T * x_i + w_0 - y_i)^2
        _MSE += np.power(np.dot(w,x[:,i])-y[i],2)
    return _MSE/n
```

```python
# COMPUTE WEIGHT VECTOR FOR LAMBDA = 100

# lambda parameter
_lambda = 100

# add bias row to training data
sXbias = np.vstack([np.ones(n),spheredX])

# compute w = [Phi^T * Phi - lambda * modified_id]^{-1} * Phi * y
w = compute_w(_lambda,sXbias,ytrain)

# print w_1,...,w_5
for i in range(1,6):
    print("w_" + str(i), "=", w[i])
```

```python
# COMPUTE WEIGHT VECTORS FOR LAMBDA RANDING FROM 0 TO 10^5

# weight matrix where W[i] corresponds to the best fit w for the i-th value of lambda
W = np.zeros((200,d+1))
# lambda vector with the values of lambda as entries
Lambda = np.linspace(0,np.power(10,5),200)
#print(Lambda)


# compute Lambda matrix and best fit w for each lambda
for i in range(200):
    #print(i)
    W[i] = compute_w(Lambda[i],sXbias,ytrain)

# vector of MSE for each lambda
MSE = np.zeros(200)
# compute MSE on training data for each lambda
for i in range(200):
    MSE[i] = mse(W[i],sXbias,ytrain)

    
# plot weights on each feature as a function of lambda
for i in range(d):
    # use log scale to see the evolution more clearly
    plt.yscale("symlog")
    plt.xscale("symlog")
    plt.plot(Lambda,W[:,i+1],label = feature_names[i])

lgd = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',ncol=4, borderaxespad=0.)

plt.title('Weight Trajectories for Ridge Descent')
plt.xlabel('lambda')
plt.ylabel('Weight')


# save plot
plt.savefig("weight_plot_p1.png",bbox_extra_artists=(lgd,), bbox_inches='tight')

plt.show()
    
```

```python
# PLOT MSE VS REGULARIZATION PARAMETER LAMBDA

plt.plot(Lambda,MSE)

plt.title('Mean Squared Error for each lambda')
plt.xlabel('lambda')
plt.ylabel('MSE')

plt.savefig("mse_p1_c.png")

plt.show()
```

```python
# COMPUTE WEIGHT VECTORS FOR LAMBDA RANDING FROM 1 TO 100

# weight matrix
W2 = np.zeros((200,d+1))
# regularization parameters
Lambda2 = np.linspace(1,100,200)
# test mean squared errors
MSE_test = np.zeros(200)
# training mean squared errors
MSE_train = np.zeros(200)

# shape of test data
dt,nt = np.shape(Xtest)

# sphere test data
sXtest = np.zeros((dt,nt))
for i in range(d):
    # sphered values of i-th feature of the x instances
    sXtest[i] = (Xtest[i] - mean[i]*np.ones(nt))/sd[i]
sXbias_test = np.vstack([np.ones(nt),sXtest])


# compute regression weights for the different lambdas
for i in range(200):
    W2[i] = compute_w(Lambda2[i],sXbias,ytrain)

# compute MSE for each found weight
for i in range(200):
    MSE_train[i] = mse(W2[i],sXbias,ytrain)
    MSE_test[i] = mse(W2[i],sXbias_test,ytest)


# plot train MSE and test MSE vs the regularization parameter lambda
plt.plot(Lambda2,MSE_train, label = "Train MSE")
plt.plot(Lambda2,MSE_test, label = "Test MSE")
plt.legend()

plt.title('Train and Test Mean Squared Error for lambda Ranging from 1 to 100')
plt.xlabel('lambda')
plt.ylabel('MSE')

plt.savefig("mse_p1_d.png")

plt.show()

```

```python
# FIND BEST TRAIN AND TEST LAMBDA

index_test = np.argmin(MSE_test)
index_train = np.argmin(MSE_train)

lambda_train = Lambda2[index_train]
lambda_test = Lambda2[index_test]

print("Best train lambda:",lambda_train,"\nBest test lambda:",lambda_test)
```
